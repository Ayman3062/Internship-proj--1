```python
from sklearn.linear_model import LinearRegression
```

```python
lm = LinearRegression()
lm
```

```python
X = wd[['Rainfall']]
Y = wd['Evaporation']
```

```python
lm.fit(X,Y)
```

```python
Yhat=lm.predict(X)
Yhat[0:5]
```

```python
import seaborn as sns
sns.regplot(x=X,y=Y,data=wd)
```
